# Solve

```python
uniq_nop_main = set(df_main.nop.unique())
uniq_nop_sim = set(df_sim.nop.unique())
```

```python
## Exist in Sim but not Main --> belom terpetakan
nop_bt = uniq_nop_sim - uniq_nop_main
len(nop_bt)
```

```python
## Exist in both
len(uniq_nop_main & uniq_nop_sim)
```

```python
## yang nop null di main bisa ada di "belom terpetakan"
```

```python
## belom terpetakan --> yang mungkin bisa isi null
df_sim_bt = df_sim[df_sim.nop.isin(nop_bt)]
df_sim_bt.sample(5)
```

```python
## df main nop is null
df_main_null = df_main[df_main.nop.isnull()]
df_main_null.sample(5)
```

```python
col_join_mapping = {'rw':'rw_op', 'rt':'rt_op', 'nop':'nop_old'}
df_main_null = df_main_null.rename(columns = col_join_mapping)
```

```python
# col_join_use = ['rw_op','rt_op','alamat_op', 'luas_bumi']
col_join_use = ['rw_op','rt_op','alamat_op']

for c in col_join_use:
    df_main_null[c] = df_main_null[c].astype(str)
    df_main_null[c] = df_main_null[c].str.strip()
    df_sim_bt[c] = df_sim_bt[c].astype(str)
    df_sim_bt[c] = df_sim_bt[c].str.strip()

map_rtrw = lambda x : str(int(x[-3:]))
## remove RT / RW
df_sim_bt['rw_op'] = df_sim_bt['rw_op'].apply(map_rtrw)
df_sim_bt['rt_op'] = df_sim_bt['rt_op'].apply(map_rtrw)
```

```python
df_join = pd.merge(df_main_null, df_sim_bt[col_join_use+['nop']], on = col_join_use, how = 'left')
df_join[~(df_join['nop'].isnull())]
```
